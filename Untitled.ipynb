```python
import pandas as pd
```

```python
df = pd.read_csv("solicitudes_credito.csv", sep=";")
dftmp = df.copy()
```

```python
display(
    dftmp,
    dftmp.dtypes
)
```

```python
# Limpieza de campo 'Sexo'
dftmp.sexo = dftmp.sexo.str.lower()
dftmp.sexo = dftmp.sexo.astype('category')
display(dftmp.sexo)
```

```python
# Limpieza de campo 'Tipo de Emprendimiento'
dftmp.tipo_de_emprendimiento = dftmp.tipo_de_emprendimiento.str.lower()
dftmp.tipo_de_emprendimiento = dftmp.tipo_de_emprendimiento.astype('category')
display(dftmp.tipo_de_emprendimiento)
```

```python
# Categorizar columna 'Estrato'
dftmp.estrato = dftmp.estrato.astype('category')
display(dftmp.estrato)
```

```python
# Limpiar columna 'Fecha de beneficio'
from dateutil.parser import parse

display(dftmp.fecha_de_beneficio)
dftmp.fecha_de_beneficio = dftmp.fecha_de_beneficio.apply(parse)
display(dftmp.fecha_de_beneficio)
```

```python
# Limpiar columna 'Monto de beneficio'
display(dftmp.monto_del_credito)
dftmp.monto_del_credito = dftmp.monto_del_credito.apply(lambda x: float(x.strip('$').replace(',', '')))
display(dftmp.monto_del_credito)
```

```python
# Limpieza de campo 'Línea de credito'
display(dftmp.línea_credito)
dftmp.línea_credito = dftmp.línea_credito.str.lower()
dftmp.línea_credito = dftmp.línea_credito.astype('category')
display(dftmp.línea_credito)
```
